# CAMUS 2D DATABASE VISUALIZATION

```python
import skimage.io as io
import matplotlib.pyplot as plt
import numpy as np
import zipfile
```

## Patient 0001 - Testing datase (50 patients)

- Two-chamber (2CH) view acquisition for both End Diastole (ED) and End Systolic (ES).

- Four-chamber (4CH) view acquisition for both End Diastole (ED) and End Systolic (ES).

```python
path = "testing/patient0001/patient0001_2CH_ED.mhd"
path2 = "testing/patient0001/patient0001_2CH_ES.mhd"
path3 = "testing/patient0001/patient0001_2CH_sequence.mhd"

img = io.imread(path, plugin='simpleitk')
img2 = io.imread(path2, plugin='simpleitk')
img3 = io.imread(path3, plugin='simpleitk')

plt.figure(figsize = (15, 4))

plt.subplot(1, 3, 1)
plt.imshow(img[0], cmap = "gray")
plt.title("2CH_ED")
plt.axis('auto')
print("Image 1 shape: ", img.shape)

plt.subplot(1, 3, 2)
plt.imshow(img2[0], cmap = "gray")
plt.title("2CH_ES")
plt.axis('auto')
print("Image 2 shape: ", img2.shape)

plt.subplot(1, 3, 3)
plt.imshow(img3[0], cmap = "gray")
plt.title("2CH_sequence")
plt.axis('auto')
print("Image 3 shape: ", img3.shape)

plt.show()
```

```python
path4 = "testing/patient0001/patient0001_4CH_ED.mhd"
path5 = "testing/patient0001/patient0001_4CH_ES.mhd"
path6 = "testing/patient0001/patient0001_4CH_sequence.mhd"

img4 = io.imread(path4, plugin='simpleitk')
img5 = io.imread(path5, plugin='simpleitk')
img6 = io.imread(path6, plugin='simpleitk')

plt.figure(figsize = (15, 5))

plt.subplot(1, 3, 1)
plt.imshow(img4[0], cmap = "gray")
plt.title("4CH_ED")
plt.axis('auto')
print("Image 4 shape: ", img4.shape)

plt.subplot(1, 3, 2)
plt.imshow(img5[0], cmap = "gray")
plt.title("4CH_ES")
plt.axis('auto')
print("Image 5 shape: ", img5.shape)

plt.subplot(1, 3, 3)
plt.imshow(img6[3], cmap = "gray")
plt.title("4CH_sequence")
plt.axis('auto')
print("Image 6 shape: ", img6.shape)

plt.show()
```

## Patient 0001 - Training dataset (500 patients with ground truth)

```python
path = "training/patient0001/patient0001_2CH_ED.mhd"
path2 = "training/patient0001/patient0001_2CH_ED_gt.mhd"
path3 = "training/patient0001/patient0001_2CH_ES.mhd"
path4 = "training/patient0001/patient0001_2CH_ES_gt.mhd"
path5 = "training/patient0001/patient0001_2CH_sequence.mhd"

img = io.imread(path, plugin='simpleitk')
img2 = io.imread(path2, plugin='simpleitk')
img3 = io.imread(path3, plugin='simpleitk')
img4 = io.imread(path4, plugin='simpleitk')
img5 = io.imread(path5, plugin='simpleitk')

plt.figure(figsize = (15, 10))

plt.subplot(2, 3, 1)
plt.imshow(img[0], cmap = "gray")
plt.title("2CH_ED")
plt.axis('auto')

plt.subplot(2, 3, 2)
plt.imshow(img2[0], cmap = "gray")
plt.title("2CH_ED_gt")
plt.colorbar()
plt.axis('auto')

plt.subplot(2, 3, 3)
plt.imshow(img5[0], cmap = "gray")
plt.title("2CH_sequence")
plt.axis('auto')

plt.subplot(2, 3, 4)
plt.imshow(img3[0], cmap = "gray")
plt.title("2CH_ES")
plt.axis('auto')

plt.subplot(2, 3, 5)
plt.imshow(img4[0], cmap = "gray")
plt.title("2CH_ES_gt")
plt.colorbar()
plt.axis('auto')
print("Image 4 shape: ", img4.shape)

plt.show()
```
