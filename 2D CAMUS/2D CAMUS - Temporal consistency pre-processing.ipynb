# TED 2D CAMUS - Temporal consistency pre-processing

```python
# Imports
from skimage import io
import os
import matplotlib.pyplot as plt
import numpy as np
import SimpleITK as sitk
```

# imagesTr and labelsTr generation

```python
# Loop first 30 patients
for i in range(1, 31):
    
    # Patient010
    if (i > 9):
        # Load mhd image
        img = sitk.ReadImage('D:/TFG/TED/database/patient0{}/patient0{}_4CH_sequence.mhd'.format(i, i))

        # Get raw image data
        img_data = sitk.GetArrayFromImage(img)
        
        # labelsTr
        # Load mhd image
        img_gt = sitk.ReadImage('D:/TFG/TED/database/patient0{}/patient0{}_4CH_sequence_gt.mhd'.format(i, i))

        # Get raw image data
        img_data_gt = sitk.GetArrayFromImage(img_gt)

        # Loop frames
        for j in range(1, img.GetSize()[2]-1, 2):

            # Reshape data to add new dimension of size 1
            img_data_previous = img_data[j-1].reshape((1,) + img_data[0].shape)
            img_data_current = img_data[j].reshape((1,) + img_data[0].shape)
            img_data_next = img_data[j+1].reshape((1,) + img_data[0].shape)

            # Convert back to sitk image
            img_previous = sitk.GetImageFromArray(img_data_previous)
            img_current = sitk.GetImageFromArray(img_data_current)
            img_next = sitk.GetImageFromArray(img_data_next)
            
            # labelsTr
            img_data_current_gt = img_data_gt[j].reshape((1,) + img_data_gt[0].shape)
            img_current_gt = sitk.GetImageFromArray(img_data_current_gt)
            
            # Save modified image
            if (j > 9):
                # imagesTr
                sitk.WriteImage(img_previous, 'D:/TFG/TED/Task550_2DLeftVentricle/imagesTr/patient0{}_{}_000{}.nii'.format(i, j, '0'))
                sitk.WriteImage(img_current, 'D:/TFG/TED/Task550_2DLeftVentricle/imagesTr/patient0{}_{}_000{}.nii'.format(i, j, '1'))
                sitk.WriteImage(img_next, 'D:/TFG/TED/Task550_2DLeftVentricle/imagesTr/patient0{}_{}_000{}.nii'.format(i, j, '2'))
                
                # labelsTr
                sitk.WriteImage(img_current_gt, 'D:/TFG/TED/Task550_2DLeftVentricle/labelsTr/patient0{}_{}.nii'.format(i, j))

            else:
                sitk.WriteImage(img_previous, 'D:/TFG/TED/Task550_2DLeftVentricle/imagesTr/patient0{}_0{}_000{}.nii'.format(i, j, '0'))
                sitk.WriteImage(img_current, 'D:/TFG/TED/Task550_2DLeftVentricle/imagesTr/patient0{}_0{}_000{}.nii'.format(i, j, '1'))
                sitk.WriteImage(img_next, 'D:/TFG/TED/Task550_2DLeftVentricle/imagesTr/patient0{}_0{}_000{}.nii'.format(i, j, '2'))
                
                 # labelsTr
                sitk.WriteImage(img_current_gt, 'D:/TFG/TED/Task550_2DLeftVentricle/labelsTr/patient0{}_0{}.nii'.format(i, j))
                
    else:
        # imagesTr
        # Load mhd image
        img = sitk.ReadImage('D:/TFG/TED/database/patient00{}/patient00{}_4CH_sequence.mhd'.format(i, i))

        # Get raw image data
        img_data = sitk.GetArrayFromImage(img)
        
        # labelsTr
        # Load mhd image
        img_gt = sitk.ReadImage('D:/TFG/TED/database/patient00{}/patient00{}_4CH_sequence_gt.mhd'.format(i, i))

        # Get raw image data
        img_data_gt = sitk.GetArrayFromImage(img_gt)

        # Loop frames
        for j in range(1, img.GetSize()[2]-1, 2):
            
            # imagesTr
            # Reshape data to add new dimension of size 1
            img_data_previous = img_data[j-1].reshape((1,) + img_data[0].shape)
            img_data_current = img_data[j].reshape((1,) + img_data[0].shape)
            img_data_next = img_data[j+1].reshape((1,) + img_data[0].shape)

            # Convert back to sitk image
            img_previous = sitk.GetImageFromArray(img_data_previous)
            img_current = sitk.GetImageFromArray(img_data_current)
            img_next = sitk.GetImageFromArray(img_data_next)
            
            # labelsTr
            img_data_current_gt = img_data_gt[j].reshape((1,) + img_data_gt[0].shape)
            img_current_gt = sitk.GetImageFromArray(img_data_current_gt)

            # Save modified image
            if (j > 9):
                # imagesTr
                sitk.WriteImage(img_previous, 'D:/TFG/TED/Task550_2DLeftVentricle/imagesTr/patient00{}_{}_000{}.nii'.format(i, j, '0'))
                sitk.WriteImage(img_current, 'D:/TFG/TED/Task550_2DLeftVentricle/imagesTr/patient00{}_{}_000{}.nii'.format(i, j, '1'))
                sitk.WriteImage(img_next, 'D:/TFG/TED/Task550_2DLeftVentricle/imagesTr/patient00{}_{}_000{}.nii'.format(i, j, '2'))
                
                # labelsTr
                sitk.WriteImage(img_current_gt, 'D:/TFG/TED/Task550_2DLeftVentricle/labelsTr/patient00{}_{}.nii'.format(i, j))

            else:
                # imagesTr
                sitk.WriteImage(img_previous, 'D:/TFG/TED/Task550_2DLeftVentricle/imagesTr/patient00{}_0{}_000{}.nii'.format(i, j, '0'))
                sitk.WriteImage(img_current, 'D:/TFG/TED/Task550_2DLeftVentricle/imagesTr/patient00{}_0{}_000{}.nii'.format(i, j, '1'))
                sitk.WriteImage(img_next, 'D:/TFG/TED/Task550_2DLeftVentricle/imagesTr/patient00{}_0{}_000{}.nii'.format(i, j, '2'))
                
                # labelsTr
                sitk.WriteImage(img_current_gt, 'D:/TFG/TED/Task550_2DLeftVentricle/labelsTr/patient00{}_0{}.nii'.format(i, j))
```
