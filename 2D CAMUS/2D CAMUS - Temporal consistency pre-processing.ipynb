```python
import os
import math

# Define the directory where the .CFG files are located
directory = 'D:/TFG/TED/cfg_files'

# Initialize a data structure to store the ED, ES, NbFrame, Sex, Age, ImageQuality, EF and LVef information for each file
file_data = {}

# Loop through all .CFG files in the directory
for filename in os.listdir(directory):
    if filename.endswith('.cfg'):
        # Open the file and extract the ED, ES, Sex, Age, ImageQuality, EF and LVef information
        with open(os.path.join(directory, filename), 'r') as file:
            data = file.read()
            ed = int(data.split('\n')[0].split(': ')[1])
            es = int(data.split('\n')[1].split(': ')[1])
            diff = abs(es-ed)
            sex = data.split('\n')[3].split(': ')[1]
            age = int(data.split('\n')[4].split(': ')[1])
            image_quality = data.split('\n')[5].split(': ')[1]
            ef_or_lvef = int(math.floor(float(data.split('\n')[6].split(': ')[1]))) if data.split('\n')[6].startswith('EF') else int(math.floor(float(data.split('\n')[8].split(': ')[1])))

            file_data[filename] = (diff, sex, age, image_quality)

# Compare the EF from one file with the LVef from another file
matches = []
for file1, data1 in file_data.items():
#     print(data1[0:])
    if file1.startswith("Info_4CH"):
        for file2, data2 in file_data.items():
#             print(data2[0:])
            if file2.startswith("patient") and file1 != file2 and data1[0:] == data2[0:]:
                matches.append((file1, file2))

# Print the matches
if len(matches) == 0:
    print('No matches found.')
else:
    print('{} matches found:'.format(len(matches)))
    sorted_list = sorted(matches, key=lambda x: x[1])
#     sorted_list = sorted(matches, key=lambda x: int(x[0].split('_')[2].split('.')[0]))
    for match in sorted_list:
        print(match)
```
